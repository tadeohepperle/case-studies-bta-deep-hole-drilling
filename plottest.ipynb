```python
import numpy as np
import matplotlib.pyplot as plt

def generate_random_time_series(n_points):
    # Generate random data points for the time series
    # You can replace np.random.randn with any other random data generation method
    random_data = np.random.randn(n_points)

    # Generate time indices (e.g., time steps, timestamps, etc.)
    time_indices = np.arange(n_points)

    return time_indices, random_data

# Number of data points in the time series
n_points = 100

# Generate the random time series
time_indices, random_data = generate_random_time_series(n_points)

# Plot the random time series
plt.figure(figsize=(10, 6))
plt.plot(time_indices, random_data)
plt.title('Random Time Series')
plt.xlabel('Time')
plt.ylabel('Value')

# Mark the segment corresponding to the Second World War (from year 20 to 40)
start_year = 20
end_year = 40
plt.axvspan(start_year, end_year, color='gray', alpha=0.3, label='Second World War')
plt.legend()
plt.grid(True)
plt.show()
```
